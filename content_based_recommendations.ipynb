## creating the corpas in list of list formate

```python
corpus = [dictionary.doc2bow(movie) for movie in word_list]
```

```python
print(corpus)
```

```python
# bag of words for 3rd movie
print(corpus[2])
```

## TF-IDF creation

```python
from gensim.models.tfidfmodel import TfidfModel
```

```python
tfidf = TfidfModel(corpus)
```

```python
tfidf[corpus[2]]
```

## Generate Similarity Matrix

```python
from gensim.similarities import MatrixSimilarity
```

```python
sim = MatrixSimilarity(tfidf[corpus],num_features=len(dictionary))
```

```python
sim[corpus[1]]
```

## creating modular code as a function

```python
def query_recommendation(movie_title, number_of_hits=5):
    movie = movies.loc[movies.primaryTitle==movie_title]
    keywords = movie['keywords'].iloc[0].split(',')    
    query_doc = keywords
    
    query_doc_bow = dictionary.doc2bow(query_doc)
    query_doc_tfidf = tfidf[query_doc_bow] 

    similarity_array = sim[query_doc_tfidf]

    similarity_series = pd.Series(similarity_array.tolist(), index=movies.primaryTitle.values)
    recommended_movies = similarity_series.sort_values(ascending=False)[1:number_of_hits+1] 
    print("Top Recommended Movies and Similarity Score: ")
    print(recommended_movies)
```

```python
query_recommendation('Batman Begins', 5)
```
